```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('/content/universal_top_spotify_songs.csv')
country_codes_df = pd.read_csv('/content/iso_country_codes.csv')
```

```python
# Print the total number of records
total_records = df.shape[0]
print(f"Total number of records: {total_records}")
```

```python
# mapping dictionary from country_codes_df
country_code_to_name = dict(zip(country_codes_df['Code'], country_codes_df['Name']))

# function for mapping country names
def get_country_name(code):
    # Check if the code is NaN or null, from dataset docs: If Null, then the country playlist is 'Global Top 50'."
    if pd.isna(code):
        return "Global"
    else:
        return country_code_to_name.get(code, "Unknown")  # count of "Unknown" in dataset as of 12/19 is 0, leaving here just in case ...

# apply the function to the country column
df['country_name'] = df['country'].apply(get_country_name)
```

```python
df
```

```python
# Convert 'snapshot_date' column to datetime and filter for today's date
df['snapshot_date'] = pd.to_datetime(df['snapshot_date'])
today_date = datetime(2023, 12, 19)
df_today = df[df['snapshot_date'] == today_date]

# Sort the DataFrame by country and rank
df_today_sorted = df_today.sort_values(by=['country_name', 'daily_rank'])

# Group by country, then select the top song for each country
number_one_songs_today = df_today_sorted.groupby('country_name').first()

number_one_songs_today
```

```python
# Filter by country example
df_filter_country = df[df['country_name'] == 'Canada']

# Sort by date and rank
df_filter_sorted = df_filter_country.sort_values(by=['snapshot_date', 'daily_rank'])

# Group by date, select the top song for each date, display specific columns
number_one_songs_by_country = df_filter_sorted.groupby('snapshot_date').first()[['country_name', 'artists', 'name']]
number_one_songs_by_country
```
